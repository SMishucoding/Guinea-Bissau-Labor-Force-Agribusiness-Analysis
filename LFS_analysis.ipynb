```python
def classify_agri_food(row):
    sect = row["TP4_SECT"]
    bran = row["TP13_BRAN"]

    if sect == 1:
        return "Primary agriculture"

    if pd.isna(bran):
        return "Non-agriculture"

    code = int(bran)

    if code in [2012, 2021, 2821, 4653]:
        return "Upstream inputs"
    elif 1000 <= code < 1100:
        return "Food manufacturing"
    elif code in [4620, 4630, 4721, 4781]:
        return "Distribution"
    elif (4900 <= code < 5000) or code == 5210:
        return "Agri-services"
    else:
        return "Non-agriculture"

df["agri_value_chain"] = df.apply(classify_agri_food, axis=1)

df = df[df["agri_value_chain"] != "Non-agriculture"].copy()

df["sector_group"] = np.where(
    df["agri_value_chain"] == "Primary agriculture",
    "Agriculture",
    "Agribusiness"
)

df["gender"] = df["sexe"].map({1: "Male", 0: "Female"})

df["agri_value_chain"].value_counts()
```

Education Analysis

```python
df["nivinst"].value_counts(dropna=False)
```

```python
def map_education(x):
    if x == 0:
        return "No education"
    elif x == 1:
        return "Primary"
    elif x == 2:
        return "Secondary"
    elif x == 3:
        return "Tertiary"
    else:
        return "Unknown"

df["education_group"] = df["nivinst"].apply(map_education)
```

```python
df["education_group"].value_counts(dropna=False)
```

```python
pd.crosstab(df["sector_group"], df["education_group"], normalize="index") * 100
```

| Sector       | No education | Primary | Secondary | Tertiary |
| ------------ | -----------: | ------: | --------: | -------: |
| Agribusiness |        17.7% |   62.9% |     17.7% |     1.6% |
| Agriculture  |        46.2% |   47.8% |      5.0% |     0.9% |


Agriculture has a much lower education profile: nearly half of workers have no education, and only about 6% have secondary or tertiary education. Agribusiness workers appear more educated, with a much lower share of workers with no education and a higher share with secondary education. This suggests that access to agribusiness jobs may require higher levels of human capital than primary agriculture, even though most agribusiness workers still have only primary education.

Formality by Sector

```python
df["formal"] = np.where(
    (df["AP6BA"] == 1) |
    (df["AP6BB"] == 1) |
    (df["AP6BC"] == 1) |
    (df["AP8C2"] == 1),
    "Formal",
    "Informal"
)
```

```python
pd.crosstab(
    [df["sector_group"], df["gender"]],
    df["formal"],
    normalize="index"
) * 100
```

Formal employment is defined using a composite indicator based on enterprise and job characteristics. A worker is considered formal if they are employed in a registered business, possess a tax identification number, are covered by social security, or have a formal work contract. This approach captures multiple dimensions of formality and is appropriate in contexts where single indicators may understate formal employment.

| Group                 | Formal | Informal |
| --------------------- | ------ | -------- |
| Agribusiness – Female | 20.6%  | 79.4%    |
| Agribusiness – Male   | 10.7%  | 89.3%    |
| Agriculture – Female  | 4.3%   | 95.7%    |
| Agriculture – Male    | 1.0%   | 99.0%    |


The analysis reveals extremely high levels of informality across both agriculture and agribusiness. Primary agriculture is almost entirely informal, with over 95 percent of workers lacking formal employment arrangements. While agribusiness shows relatively higher levels of formality, the majority of jobs remain informal. Interestingly, women appear more likely to be in formal employment compared to men, particularly in agribusiness; however, this likely reflects their concentration in specific segments such as small-scale trade rather than broad-based access to high-quality formal jobs. Overall, the findings highlight informality as a pervasive structural feature of the labor market.
